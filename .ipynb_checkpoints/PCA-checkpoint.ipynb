```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import LeaveOneOut
from mpl_toolkits import mplot3d
#import libs
```

```python
####导入数据

file_name = 'C:/Users/86180/Desktop/features.xlsx'
data = pd.read_excel(file_name)
#data = pd.read_csv('digit.csv')
x = data.iloc[:,1:]
y = data.iloc[:,0]
print(x.shape)
data.head()
```

```python
####画累计方差贡献率曲线，找最佳降维后维度的范围u
pca = PCA().fit(x)  #默认参数，什么都不填数据最小维度（行42000、列784）————》784
plt.figure(figsize=(20,6))
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('降维后的特征数目')
plt.ylabel('累计可解释性方差比例之和')
plt.show()
```

```python
audiometric = x
```

```python
pd.options.display.max_columns = 10
round(audiometric.corr(), 2)
#calculate rho
sns.heatmap(round(audiometric.corr(), 2),annot=True)
#plot heatmap
```

```python
scaler = StandardScaler()
scaler.fit(audiometric)
X = scaler.transform(audiometric)
#normalize the data
```

```python
model = PCA()
model.fit(X)
#variance can be explained
model.explained_variance_
#percentage of variance can be explained
model.explained_variance_ratio_
#plot
plt.plot(model.explained_variance_ratio_, 'o-')
plt.xlabel('Principal Component')
plt.ylabel('Proportion of Variance Explained')
plt.title('PVE')
```

```python
#Cumulative PVE
plt.plot(model.explained_variance_ratio_.cumsum(), 'o-')
plt.xlabel('Principal Component')
plt.ylabel('Cumulative Proportion of Variance Explained')
plt.axhline(0.9, color='k', linestyle='--', linewidth=1)
plt.title('Cumulative PVE')
```

```python
#主成分核载矩阵
model.components_
 
columns = ['PC' + str(i) for i in range(1, 7)]
 
pca_loadings = pd.DataFrame(model.components_, columns=audiometric.columns, index=columns)
round(pca_loadings, 2)
```
